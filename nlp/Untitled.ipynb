```python
from nltk import SnowballStemmer

def my_tokenizer(doc, tkpat=re.compile('\\b[a-z][a-z][a-z]+\\b')):
    stemmer = SnowballStemmer('english')
    return [stemmer.stem(token) for token in tkpat.findall(doc)]
```

```python
#text = ['That it should come to this!', 'This above all: to thine own self be true.', 'Something is rotten in the state of Denmark.']

# CountVectorizer is a class; so `vectorizer` below represents an instance of that object.
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=0.05,ngram_range=(1,1), analyzer ='word', input = 'filename',\
                             stop_words='english', tokenizer = my_tokenizer)

corpus = file_list
# call `fit` to build the vocabulary
tf_vectorizer.fit(corpus)

tf_vector = tf_vectorizer.transform(corpus)
```
